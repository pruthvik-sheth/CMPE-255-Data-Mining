```python
# Cell 1: Load FIFA 21 Player Data
import pandas as pd

# Load the latest dataset (players_21.csv)
fifa_data = pd.read_csv('players_21.csv')

# Display the first few rows of the dataset
fifa_data.head()
```

```python
# Cell 2: Generate Sweetviz Report
import sweetviz as sv

# Analyze the dataset
fifa_report = sv.analyze(fifa_data)

# Save the report to an HTML file
fifa_report.show_html("fifa21_sweetviz_report.html")

print("Sweetviz report generated and saved as 'fifa21_sweetviz_report.html'.")
```
